```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# 1. Geração de Dados Sintéticos (Simulando pacientes)
print("Gerando dados sintéticos de pacientes...")

np.random.seed(42) # Para reprodutibilidade

data = {
    'Idade': np.random.randint(20, 80, 100),
    'Sintoma_1_Escala': np.random.randint(0, 11, 100), # Escala de 0 a 10
    'Resultado_Exame_Valor': np.random.uniform(5.0, 30.0, 100) # Valor contínuo
}
df = pd.DataFrame(data)

print("\nPrimeiras 5 linhas do DataFrame:")
print(df.head())
print("-" * 50)

# 2. Pré-processamento dos Dados
# É crucial padronizar os dados para K-Means, pois ele é baseado em distância.
# Standard Scaler (Z-score) transforma os dados para média=0 e desvio padrão=1.
features = ['Idade', 'Sintoma_1_Escala', 'Resultado_Exame_Valor']
X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Dados padronizados (primeiras 5 linhas):")
print(X_scaled[:5])
print("-" * 50)

# 3. Determinação do Número Ideal de Clusters (K)
# O "Método do Cotovelo" (Elbow Method) é frequentemente usado para escolher K.
# Ele calcula a Soma dos Quadrados das Distâncias (Inertia) para diferentes valores de K.
# O "cotovelo" na curva indica um bom K.

wcss = [] # Within-Cluster Sum of Squares (Inertia)
max_k = 10
for i in range(1, max_k + 1):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plotagem do Método do Cotovelo
plt.figure(figsize=(8, 5))
plt.plot(range(1, max_k + 1), wcss, marker='o', linestyle='--')
plt.title('Método do Cotovelo (Elbow Method)')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('WCSS / Inertia')
plt.grid(True)
plt.show()

# Baseado no gráfico (e para fins práticos do exemplo), vamos escolher K=3

# 4. Aplicação do Algoritmo K-Means
# Definindo K=3 (Exemplo)
K_escolhido = 3
kmeans_final = KMeans(n_clusters=K_escolhido, init='k-means++', random_state=42, n_init=10)

# Treinamento e atribuição dos rótulos de cluster
df['Cluster'] = kmeans_final.fit_predict(X_scaled)

# 5. Avaliação do Modelo (Métrica Silhueta)
# A Silhueta mede o quão bem um objeto se encaixa em seu próprio cluster
# em comparação com clusters vizinhos. O valor varia de -1 a 1.
# Próximo de 1: O objeto está bem agrupado.
# Próximo de 0: Os clusters se sobrepõem.
# Próximo de -1: O objeto foi agrupado no cluster errado.

score_silhueta = silhouette_score(X_scaled, df['Cluster'])

print(f"**Resultado Final do K-Means (K={K_escolhido})**")
print("-" * 50)
print(f"Métrica Silhueta: **{score_silhueta:.4f}**")
print("-" * 50)

# 6. Análise dos Clusters Formados
print("\nAnálise Estatística dos Clusters:")
cluster_analysis = df.groupby('Cluster')[features].mean()
print(cluster_analysis)
print("\nInterpretando: O Cluster 0 tem pacientes de X anos, com Y de sintomas e Z de exames, etc.")

# 7. Visualização 3D (Opcional, mas útil para 3 variáveis)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Mapeando os clusters para cores
scatter = ax.scatter(df['Idade'], df['Sintoma_1_Escala'], df['Resultado_Exame_Valor'],
                     c=df['Cluster'], cmap='viridis', s=50)

ax.set_xlabel('Idade')
ax.set_ylabel('Sintoma 1 (Escala)')
ax.set_zlabel('Resultado do Exame')
ax.set_title(f'Agrupamento de Pacientes com K-Means (K={K_escolhido})')

# Adicionando a legenda
legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend1)
plt.show()
```
